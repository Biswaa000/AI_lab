```python
import numpy as np 
import matplotlib.pyplot as plt 
```

```python
m=0.7 
c=0.3

X=np.arange(0,1,0.02)
y=m*X+c
print(X.shape,y.shape)
```

```python
split=int(0.8*len(X))
X_train,y_train=X[:split],y[:split]
X_test,y_test=X[split:],y[split:]
```

```python
def plot_data(X_train=X_train,y_train=y_train,X_test=X_test,y_test=y_test,prediction=None):
    plt.figure(figsize=(8,6))
    plt.scatter(X_train,y_train,c='b',s=8,label="training data")
    plt.scatter(X_test,y_test,c='r',s=8,label="testing data")

    if prediction is not None:
        plt.scatter(X_test,prediction,c='g',s=8,label="testing data")
    plt.legend()
    
plot_data()
```

```python
W= np.random.randn(1)
b=np.random.randn(1)
print(W,b)
```

```python
def loss_fn(y_actual,y_predictions):
    return np.sum(np.square((y_actual-y_predictions)))/len(y_actual)
```

```python
# Learning rate and number of epochs
lr = 0.1
epochs = 300

# Lists to store loss values
train_loss_values = []
test_loss_values = []

# # Initialize weights (W) and bias (b)
# W = np.random.randn()  # Random initial weight
# b = np.random.randn()  # Random initial bias


# Training loop
for epoch in range(epochs):
    # Predicted values for the training set
    y_pred = X_train * W + b

    # Calculate training loss
    train_loss = loss_fn(y_actual=y_train, y_predictions=y_pred)
    train_loss_values.append(train_loss)

    # Compute gradients
    gradientW = (-2 * np.sum((y_train - y_pred) * X_train)) / len(y_pred)
    gradientb = (-2 * np.sum(y_train - y_pred)) / len(y_pred)

    # Update weights and bias
    W = W - lr * gradientW
    b = b - lr * gradientb

    # Predicted values for the test set
    test_pred = X_test * W + b

    # Calculate test loss
    test_loss = loss_fn(y_actual=y_test, y_predictions=test_pred)
    test_loss_values.append(test_loss)

    # Print progress every 50 epochs
    if epoch % 50 == 0:
        print(f"Epoch {epoch}, Train Loss: {train_loss:.4f}, Test Loss: {test_loss:.4f}")

# Final values
print(f"Final W: {W}, Final b: {b}")
```

```python
import matplotlib.pyplot as plt

# Plot training and test loss values
plt.figure(figsize=(10, 8))
plt.plot(range(epochs), train_loss_values, label='Train Loss', color='blue', linewidth=2)
plt.plot(range(epochs), test_loss_values, label='Test Loss', color='orange', linewidth=2)

# Add labels, title, and legend
plt.xlabel('Epochs', fontsize=14)
plt.ylabel('Loss', fontsize=14)
plt.title('Training and Test Loss over Epochs', fontsize=16)
plt.legend(fontsize=12)

# Add grid for better visualization
plt.grid(True, linestyle='--', alpha=0.6)

# Show the plot
plt.show()
```
