```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
X=1.5
y=0.5

# y=m*x+c
```

```python
# W=np.random.randn(2,3)
W=np.random.randn(1)
print(W,W.shape )
```

```python
# building the training loop 
epochs=20 
lr=0.1
eList=[]
for epoch in range(epochs):
    #forward pass 
    y_out=X*W
    #compute error
    error =np.square(y-y_out)
    eList.append(error)
    #compute the gradient
    gradient=-2*(y-y_out)*X

     # backpropagate the error
    W=W-gradient*lr
```

```python
plt.plot(range(epochs),eList)
plt.xlabel("epochs")
plt.ylabel("erro")
plt.title("epochs uvs error ")
```

```python
def output(X,W): 
    return X*W
print(output(X,W))
```
