```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

X = np.array([[1.5, 0.625]]) 
y = 0.5  
```

```python
np.random.seed(42)
W_hidden = np.random.randn(2, 2)  # Weights from input to hidden layer (2 inputs → 2 hidden neurons)
W_output = np.random.randn(1, 2)  # Weights from hidden layer to output (2 hidden neurons → 1 output)
```

```python
epochs = 20
lr = 0.1
eList = []
```

```python

for epoch in range(epochs):
    # Forward pass
    hidden_output = np.dot(X, W_hidden.T)  # Output from the hidden layer 
    y_out = np.dot(hidden_output, W_output.T)  # Output from the output layer (linear)

    # Compute error
    error = np.square(y - y_out)
    eList.append(error[0][0])

    # Backpropagation
    # Gradient for output layer weights
    gradient_output = -2 * (y - y_out) * hidden_output

    # Gradient for hidden layer weights
    gradient_hidden = -2 * (y - y_out) * W_output * X.T

    # Update weights
    W_output -= lr * gradient_output
    W_hidden -= lr * gradient_hidden.T
```

```python
# Print final weights and gradients
print("Final Weights (Hidden Layer):\n", W_hidden)
print("Final Weights (Output Layer):\n", W_output)
```

```python
# Plot error vs. epochs
plt.plot(range(epochs), eList)
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.title("Epochs vs Error")
plt.show()
```

```python

def output(X, W_hidden, W_output):
    hidden_output = np.dot(X, W_hidden.T)
    final_output = np.dot(hidden_output, W_output.T)
    return final_output


print("Output after training:", output(X, W_hidden, W_output))
```
