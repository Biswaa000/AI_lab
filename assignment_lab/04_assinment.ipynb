```python
 def train(self, input_data, target, learning_rate=0.1, epochs=10000):
        """Trains the neural network using backpropagation."""

        for epoch in range(epochs):
            output, hidden = forward(input_data)

            # Calculate error
            output_error = target - output
            
            #Backpropagation
            d_output = output_error * sigmoid_derivative(output)
            
            output_error_hidden = np.dot(self.weights_ho.T, d_output)
            d_hidden = output_error_hidden * self.sigmoid_derivative(hidden)

            # Update weights and biases
            self.weights_ho += learning_rate * np.dot(d_output, hidden.T)
            self.bias_o += learning_rate * d_output
            self.weights_ih += learning_rate * np.dot(d_hidden, input_data.T)
            self.bias_h += learning_rate * d_hidden

            if epoch % 1000 == 0:
                loss = np.mean(np.abs(output_error))
                print(f"Epoch: {epoch}, Loss: {loss}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Inputs and target output
X = np.array([[1.5, 0.625]])  # Shape: (1, 2)
y = 0.5  # Target output

# Initialize weights
np.random.seed(42)
W_hidden = np.random.randn(2, 2)  # Weights from input to hidden layer (2 inputs → 2 hidden neurons)
W_output = np.random.randn(1, 2)  # Weights from hidden layer to output (2 hidden neurons → 1 output)

# Sigmoid activation function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Derivative of sigmoid
def sigmoid_derivative(x):
    return sigmoid(x) * (1 - sigmoid(x))

# Training loop parameters
epochs = 20
lr = 0.1
eList = []

for epoch in range(epochs):
    # Forward pass
    hidden_input = np.dot(X, W_hidden.T)  # Input to hidden layer
    hidden_output = sigmoid(hidden_input)  # Output of hidden layer

    final_input = np.dot(hidden_output, W_output.T)  # Input to output layer
    y_out = final_input  # Output layer (linear activation)

    # Compute error
    error = np.square(y - y_out)
    eList.append(error[0][0])

    # Backpropagation
    # Gradient for output layer
    output_gradient = -2 * (y - y_out)  # dError/dY_out

    # Gradient for hidden layer weights
    hidden_gradient = output_gradient * W_output * sigmoid_derivative(hidden_input)

    # Update weights
    W_output -= lr * output_gradient * hidden_output
    W_hidden -= lr * np.dot(hidden_gradient.T, X)

# Print final weights and output
print("Final Weights (Hidden Layer):\n", W_hidden)
print("Final Weights (Output Layer):\n", W_output)

# Plot error vs. epochs
plt.plot(range(epochs), eList)
plt.xlabel("Epochs")
plt.ylabel("Error")
plt.title("Epochs vs Error")
plt.show()

# Define the output function
def output(X, W_hidden, W_output):
    hidden_input = np.dot(X, W_hidden.T)
    hidden_output = sigmoid(hidden_input)
    final_input = np.dot(hidden_output, W_output.T)
    return final_input

# Test the network
print("Output after training:", output(X, W_hidden, W_output))
```
